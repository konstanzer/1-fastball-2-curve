```python
zack=pd.read_csv('greinke.csv')
```

```python
pid=pd.read_csv('PLAYERIDMAP.csv')
pid=pid[['PLAYERNAME', 'MLBID']]
pid.dropna(how='any', axis=0, inplace=True)
pid.rename(columns={'PLAYERNAME':'batter_name', 'MLBID':'batter'},
           inplace=True)
pid=pid.astype({'batter':'int64'})
pid.to_csv('player_ids.csv', index=False)
```

```python
zack.dropna(how='all', axis=1, inplace=True)
zack.drop(columns=['index'], axis=1, inplace=True)
```

```python
zack=zack.merge(pid)
```

```python
zack.sort_values(by='game_date', inplace=True)
zack.reset_index(inplace=True)
zack.drop(columns=['index'], axis=1, inplace=True)
```
